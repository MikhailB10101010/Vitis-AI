```python
import sqlite3
import pandas as pd
from pathlib import Path

current_path = Path.cwd()
db_folder_path = current_path.resolve().parent / "data"

db_name = "vineyard_1.db"
db_path = db_folder_path / db_name

if db_path.is_file():  # Проверяет, что это файл, а не папка
    print("Файл существует")
else:
    print("! Файл не существует !")
```

# Переименование столбцов

```python
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

table_name = "vineyard_features"
column_old = "elevation_status"
column_new = "elevation_GEE_USGS_30m_status"

try:
    query = f'''
        ALTER TABLE {table_name} 
        RENAME COLUMN {column_old} TO {column_new}
    '''

    cursor.execute(query)
    conn.commit()
    print("Столбец успешно переименован!")
except sqlite3.Error as e:
    print(f"Ошибка (возможно, столбца нет или версия SQLite старая): {e}")

conn.close()
```
