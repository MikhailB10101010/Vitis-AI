```python
from services.terrain_GEE_USGS import terrain_GEE_USGS

# Текст для батча

coords = [
    [56.728766, 60.587326],
    [55.7558, 37.6173], # Москва
    [59.9343, 30.3351], # Санкт-Петербург
]

# Cписок Feature, меняя местами lat и lon -> [lon, lat]
features = [ee.Feature(ee.Geometry.Point([lon, lat])) for lat, lon in coords]

fc = ee.FeatureCollection(features)

ans = terrain_GEE_USGS(fc)
print(ans)
```
