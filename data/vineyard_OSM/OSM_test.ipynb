```python
import osmnx as ox
import geopandas as gpd
import pandas as pd

# --- 1. Настройки ---
# Укажите интересующий вас регион. Это может быть название места,
# которое понимает Nominatim (геокодер OSM), или координаты bbox.
place_name = "Bordeaux, France"  # Пример: "Tuscany, Italy", "France"

# Теги для поиска: ищем объекты с тегом landuse=vineyard
tags = {"landuse": "vineyard"}
```

```python
# --- 2. Загрузка данных ---
print(f"Загружаем данные о виноградниках для региона: {place_name}...")
try:
    # Загружаем все геометрии (полигоны и точки) по заданным тегам
    gdf_vineyards = ox.features_from_place(place_name, tags=tags)
    print(f"Загружено объектов: {len(gdf_vineyards)}")
except Exception as e:
    print(f"Ошибка при загрузке данных: {e}")
    print("Попробуйте уточнить название региона или использовать bbox.")
    exit()
```
